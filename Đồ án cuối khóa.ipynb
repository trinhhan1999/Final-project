# Giới thiệu về đề tài 

**US SBA (US Small Business Administration)** được thành lập **năm 1953** trên nguyên tắc thúc đẩy và hỗ trợ các doanh nghiệp nhỏ trên **thị trường tín dụng Hoa Kỳ**. Các doanh nghiệp nhỏ là nguồn tạo việc làm chính ở **Hoa Kỳ**; do đó, thúc đẩy sự hình thành và tăng trưởng của doanh nghiệp nhỏ mang lại lợi ích xã hội bằng cách tạo cơ hội việc làm và giảm tỷ lệ thất nghiệp. Một cách mà SBA hỗ trợ các doanh nghiệp nhỏ này là thông qua chương trình bảo lãnh khoản vay được thiết kế để khuyến khích các ngân hàng cấp các khoản vay cho các doanh nghiệp nhỏ. SBA hoạt động giống như một nhà cung cấp bảo hiểm để giảm thiểu rủi ro cho ngân hàng bằng cách chấp nhận một số rủi ro thông qua việc đảm bảo một phần khoản vay. Trong trường hợp khoản vay bị vỡ nợ, SBA sẽ chi trả số tiền họ đã bảo lãnh.

Đã có nhiều câu chuyện thành công về việc các công ty khởi nghiệp nhận được bảo lãnh khoản vay **SBA** như **FedEx** và **Apple Computer**. Tuy nhiên, cũng có những câu chuyện về các doanh nghiệp nhỏ và/hoặc công ty mới thành lập đã không trả được các khoản vay do SBA bảo lãnh.

# Xác định mục tiêu

 - Bản chất của hoạt động cho vay là dựa trên sự tin tưởng về việc Khách hàng sẽ hoàn trả tiền vay. Trường hợp Khách hàng không trả được nợ sẽ dẫn đến thiệt hại về vốn cho ngân hàng. Một trong các mục đích của ngân hàng là hạn chế các thiệt hại này thông qua nhận diện các Khách hàng xấu để từ chối cho vay. Bên cạnh đó ngân hàng cũng muốn tối đa hoá dư nợ cho vay (quy mô cho vay) từ đó tăng thu nhập từ lãi vay. Điều đó cũng đòi hỏi ngân hàng không từ chối những Khách hàng có nhiều khả năng là Khách hàng tốt. 
 - Để đạt được mục đích đề ra ta cần phải dự đoán được liệu Khách hàng có trả được nợ hay không trong tương lai. Trường hợp kết quả dự đoán là "trả được nợ" thì ngân hàng có thể quyết định cho Khách hàng đó vay. Ngược lại nếu kết quả dự đoán là "không trả được nợ" thì để bảo toàn vốn ngân hàng có thể thông báo từ chối cho vay.


# Xác định cách tiếp cận giải quyết vấn đề

- Để xây dựng được mô hình dự đoán như yêu cầu, ta có thể sử dụng các phương pháp học có giám sát để huấn luyện mô hình trên dữ liệu đã có sẵn sau đó sử dụng mô hình để dự đoán cho các quan sát mới.
- Theo yêu cầu này, Các mô hình chọn xây dựng: Logistic Regression, Gradient Boosting (cụ thể là LightGBM)

Trong quá trình tạo lập dataset này thì người đánh máy có thể điền sai một số thông tin về trường RevLineCr và LowDoc. Thực tế thì giá trị của hai trường này chỉ có Y và N. Vì vậy chúng ta cần loại bỏ các giá trị bị sai.

```python
# chỉ lấy các giá Trị "Y" và "N" ở cả 2 trường RevLineCr và LowDoc
df_copy = df_copy[(df_copy['RevLineCr'] == 'Y') | (df_copy['RevLineCr'] == 'N')]
df_copy = df_copy[(df_copy['LowDoc'] == 'Y') | (df_copy['LowDoc'] == 'N')]

# RevLineCr và LowDoc: 0 = No, 1 = Yes
df_copy['RevLineCr'] = np.where(df_copy['RevLineCr'] == 'N', 0, 1)
df_copy['LowDoc'] = np.where(df_copy['LowDoc'] == 'N', 0, 1)

print(df_copy['RevLineCr'].unique())
print(df_copy['LowDoc'].unique())
```

```python

# Tạo trường mục tiêu 'Default' dựa trên MIS_Status trong đó P I F = 0 và CHGOFF = 1 để chúng tôi có thể xem những tính năng nào phổ biến trong khoản vay mặc định
df_copy['Default'] = np.where(df_copy['MIS_Status'] == 'P I F', 0, 1)
df_copy['Default'].value_counts()
```

```python
# Chuyển đổi các cột ApprovalDate  và DisbursementDate  thành giá trị ngày giờ
df_copy[['ApprovalDate', 'DisbursementDate']] = df_copy[['ApprovalDate', 'DisbursementDate']].apply(pd.to_datetime)
```

```python
# Tạo cột DaysToDisbursement tính toán số ngày trôi qua giữa DisbursementDate và ApprovalDate
df_copy['DaysToDisbursement'] = df_copy['DisbursementDate'] - df_copy['ApprovalDate']

# Thay đổi DaysToDisbursement từ timedelta64 dtype thành int64 dtype
df_copy['DaysToDisbursement'] = df_copy['DaysToDisbursement'].astype('str').apply(lambda x: x[:x.index('d') - 1]).astype('int64')
```

```python
# Tạo trường DisbursementFY cho tiêu chí lựa chọn thời gian sau này
df_copy['DisbursementFY'] = df_copy['DisbursementDate'].map(lambda x: x.year)
```

Một số liệu khác mà chúng ta muốn khám phá là liệu ngân hàng phục vụ khoản vay có ở cùng trạng thái với địa điểm kinh doanh hay không. Giả định là việc 
thanh toán khoản vay cho một doanh nghiệp ở tiểu bang khác sẽ khó khăn hơn và điều này có thể có tác động tiêu cực đến khả năng trả nợ của doanh nghiệp.


```python
# Tạo trường StateSame xác định vị trí của Business State giống với BankState
df_copy['StateSame'] = np.where(df_copy['State'] == df_copy['BankState'], 1, 0)
```

Liệu một khoản vay có được đảm bảo bằng bất động sản (sở hữu đất đai) hay không là một chỉ số rủi ro khác được thảo luận. Cơ sở lý luận cho chỉ số này là giá trị của đất đai thường đủ lớn để trang trải số tiền gốc chưa thanh toán, do đó làm giảm xác suất vỡ nợ. Vì thời hạn của khoản vay phụ thuộc vào tuổi thọ dự kiến của tài sản, các khoản vay được đảm bảo bằng bất động sản sẽ có thời hạn từ **20 năm trở lên (>= 240 tháng)** và là khoản vay duy nhất được cấp cho thời hạn dài như vậy, trong khi các khoản vay không được đảm bảo bằng bất động sản sẽ có thời hạn **dưới 20 năm (<240)**.

Một chỉ số rủi ro luôn xuất hiện trong cuộc thảo luận là nền kinh tế có thể tác động đến tỷ lệ vỡ nợ như thế nào. Các khoản cho vay doanh nghiệp nhỏ là bị ảnh hưởng bởi nền kinh tế nói chung, và nhiều khoản vay dành cho doanh nghiệp nhỏ có xu hướng vỡ nợ ngay trước và trong thời kỳ suy thoái kinh tế. Đối với các khoản vay đang hoạt động trong thời kỳ Đại suy thoái, chúng ta sẽ tạo một trường cho các khoản vay trong thời kỳ **Đại suy thoái (2007-2009)** giữa DisbursementFY và DisbursementFY cộng với term cho vay (tính bằng năm).

```python
# Lĩnh vực cho vay thế chấp bằng Bất động sản (thời hạn vay tối thiểu 20 năm)
df_copy['RealEstate'] = np.where(df_copy['Term'] >= 240, 1, 0)

# Lĩnh vực cho vay hoạt động trong cuộc Đại suy thoái (2007-2009)
df_copy['GreatRecession'] = np.where(((2007 <= df_copy['DisbursementFY']) & (df_copy['DisbursementFY'] <= 2009)) | 
                                     ((df_copy['DisbursementFY'] < 2007) & (df_copy['DisbursementFY'] + (df_copy['Term']/12) >= 2007)), 1, 0)
```

Chúng ta sẽ tạo thêm 1 trường để biết số tiền giải ngân có lớn hơn số tiền thực sự được phê duyệt hay không. Điều này sẽ có ý nghĩa thú vị bởi vì việc giải ngân thêm tiền có thể cho thấy rằng doanh nghiệp có nguy cơ vỡ nợ cao hơn. Tuy nhiên, điều này có thể tương quan với các dòng tín dụng quay vòng trong đó doanh nghiệp liên tục rút và thanh toán số dư.

```python
# Tạo trường để biểu thị nếu một số tiền lớn hơn đã được giải ngân so với những gì Ngân hàng đã phê duyệt

df_copy['DisbursedGreaterAppv'] = np.where(df_copy['DisbursementGross'] > df_copy['GrAppv'], 1, 0)
```

Thứ hai, chúng ta hạn chế khung thời gian cho các khoản vay bằng cách loại trừ các khoản giải ngân sau **năm 2010** do thời hạn của các khoản vay thường là **5 năm** trở lên.

```python
# Chúng ta sẽ sử dụng Inter Quartile Range để xử lý các giá trị ngoại lai
# Xác định giới hạn
def limit(i):
    Q1 = df_train_our[i].quantile(0.25)
    Q3 = df_train_our[i].quantile(0.75)
    IQR = Q3 - Q1
    
    # xác định giới hạn trên, dưới thông thường và giới hạn trên, dưới cực đại
    lower_limit = df_train_our[i].quantile(0.25) - (IQR * 1.5)
    lower_limit_extreme = df_train_our[i].quantile(0.25) - (IQR * 3)
    upper_limit = df_train_our[i].quantile(0.75) + (IQR * 1.5)
    upper_limit_extreme = df_train_our[i].quantile(0.75) + (IQR * 3)
    print('Lower Limit:', lower_limit)
    print('Lower Limit Extreme:', lower_limit_extreme)
    print('Upper Limit:', upper_limit)
    print('Upper Limit Extreme:', upper_limit_extreme)

# Tính phần trăm ngoại lệ từ dữ liệu  
def percent_outliers(i):
    Q1 = df_train_our[i].quantile(0.25)
    Q3 = df_train_our[i].quantile(0.75)
    IQR = Q3 - Q1
    
    # xác định giới hạn trên, dưới thông thường và giới hạn trên, dưới cực đại
    lower_limit = df_train_our[i].quantile(0.25) - (IQR * 1.5)
    lower_limit_extreme = df_train_our[i].quantile(0.25) - (IQR * 3)
    upper_limit = df_train_our[i].quantile(0.75) + (IQR * 1.5)
    upper_limit_extreme = df_train_our[i].quantile(0.75) + (IQR * 3)
    # xem phần trăm ngoại lệ trên tổng dữ liệu
    print('Lower Limit: {} %'.format(df_train_our[(df_train_our[i] >= lower_limit)].shape[0]/ df_train_our.shape[0]*100))
    print('Lower Limit Extereme: {} %'.format(df_train_our[(df_train_our[i] >= lower_limit_extreme)].shape[0]/df_train_our.shape[0]*100))
    print('Upper Limit: {} %'.format(df_train_our[(df_train_our[i] >= upper_limit)].shape[0]/ df_train_our.shape[0]*100))
    print('Upper Limit Extereme: {} %'.format(df_train_our[(df_train_our[i] >= upper_limit_extreme)].shape[0]/df_train_our.shape[0]*100))
```

## CUSBoost

CUSBoost dựa trên sự kết hợp giữa lấy mẫu dựa trên cụm và thuật toán Adaboost. Nó tách các thể hiện của lớp đa số và thiểu số khỏi tập dữ liệu gốc và phân cụm các thể hiện của lớp đa số thành k cụm bằng cách sử dụng thuật toán phân cụm k-means. Ở đây, tham số k được xác định bằng cách tối ưu hóa siêu tham số. Sau đó, việc lấy mẫu dưới mức ngẫu nhiên được thực hiện trên từng cụm đã tạo bằng cách chọn ngẫu nhiên 50% số phiên bản (nhưng có thể điều chỉnh tùy theo vấn đề miền hoặc tập dữ liệu) và loại bỏ phần còn lại. Vì phân cụm được sử dụng trước khi lấy mẫu, nên về mặt lý thuyết, thuật toán này sẽ hoạt động tốt nhất khi tập dữ liệu có khả năng phân cụm cao. Các mẫu đại diện này sau đó được kết hợp với các thể hiện của lớp thiểu số để thu được các bộ dữ liệu cân bằng. Sức mạnh của thuật toán nằm ở chỗ nó xem xét các ví dụ từ tất cả các không gian con của lớp đa số, vì phân cụm k-mean đặt mỗi cá thể vào một số cụm. Các phương pháp tương tự khác thường thất bại trong việc thu thập các đại diện thích hợp của tầng lớp đa số.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
#!pip install imbalanced-learn
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold, KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score, confusion_matrix, precision_score, recall_score, roc_curve, average_precision_score, precision_recall_curve
from sklearn.ensemble import AdaBoostClassifier
from sklearn.cluster import KMeans
import math




def cus_sampler(X_train, y_train, number_of_clusters=23, percentage_to_choose_from_each_cluster=0.50):

    selected_idx = []
    selected_idx = np.asarray(selected_idx)

    value, counts = np.unique(y_train, return_counts=True)
    minority_class = value[np.argmin(counts)]
    majority_class = value[np.argmax(counts)]

    idx_min = np.where(y_train == minority_class)[0]
    idx_maj = np.where(y_train == majority_class)[0]

    majority_class_instances = X_train[idx_maj]
    majority_class_labels = y_train[idx_maj]

    kmeans = KMeans(n_clusters=number_of_clusters)
    kmeans.fit(majority_class_instances)

    X_maj = []
    y_maj = []

    points_under_each_cluster = {i: np.where(kmeans.labels_ == i)[0] for i in range(kmeans.n_clusters)}

    for key in points_under_each_cluster.keys():

        points_under_this_cluster = np.array(points_under_each_cluster[key])
        number_of_points_to_choose_from_this_cluster = math.ceil(
            len(points_under_this_cluster) * percentage_to_choose_from_each_cluster)




        selected_points = np.random.choice(points_under_this_cluster,
                                           size=number_of_points_to_choose_from_this_cluster, replace=False)
        X_maj.extend(majority_class_instances[selected_points])
        y_maj.extend(majority_class_labels[selected_points])

        selected_idx = np.append(selected_idx,selected_points)

        selected_idx = selected_idx.astype(int)

    selected_idx = np.append(selected_idx, idx_min)
    X_sampled = np.concatenate((X_train[idx_min], np.array(X_maj)))
    y_sampled = np.concatenate((y_train[idx_min], np.array(y_maj)))

    return X_sampled, y_sampled, selected_idx
```

```python
class CUSBoostClassifier:
    def __init__(self, n_estimators, depth, number_of_clusters, percentage_to_choose_from_each_cluster):
        self.M = n_estimators
        self.depth = depth
        self.clusters = number_of_clusters
        self.pct_to_choose = percentage_to_choose_from_each_cluster

    def fit(self, X, Y):
        self.models = []
        self.alphas = []

        N, _ = X.shape
        W = np.ones(N) / N

        for m in range(self.M):
            tree = DecisionTreeClassifier(max_depth=self.depth, splitter='best')


            X_undersampled, y_undersampled, chosen_indices = cus_sampler(X,Y, self.clusters, self.pct_to_choose)


            tree.fit(X_undersampled, y_undersampled,
                     sample_weight=W[chosen_indices])

            P = tree.predict(X)

            err = np.sum(W[P != Y])

            if err > 0.5:
                m = m - 1
            if err <= 0:
                err = 0.0000001
            else:
                try:
                    if (np.log(1 - err) - np.log(err)) == 0 :
                        alpha = 0
                    else:
                        alpha = 0.5 * (np.log(1 - err) - np.log(err))
                    W = W * np.exp(-alpha * Y * P)  # vectorized form
                    W = W / W.sum()  # normalize so it sums to 1
                except:
                    alpha = 0
                    W = W * np.exp(-alpha * Y * P)  # vectorized form
                    W = W / W.sum()  # normalize so it sums to 1

                self.models.append(tree)
                self.alphas.append(alpha)

    def predict(self, X):
        N, _ = X.shape
        FX = np.zeros(N)
        for alpha, tree in zip(self.alphas, self.models):
            FX += alpha * tree.predict(X)
        return np.sign(FX), FX

    def predict_proba(self, X):
        proba = sum(tree.predict_proba(X) * alpha for tree , alpha in zip(self.models,self.alphas) )


        proba = np.array(proba)


        proba = proba / sum(self.alphas)

        proba = np.exp((1. / (2 - 1)) * proba)
        normalizer = proba.sum(axis=1)[:, np.newaxis]
        normalizer[normalizer == 0.0] = 1.0
        # proba =  np.linspace(proba)
        # proba = np.array(proba).astype(float)
        proba = proba /  normalizer

        return proba
```
